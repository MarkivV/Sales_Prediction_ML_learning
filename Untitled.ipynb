```python
def plot_polynomial_regression(feature, degree=2):
    X = df[[feature]]
    y = df['sales']

    # Create Polynomial Features
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    # Train the Model
    model = LinearRegression()
    model.fit(X_poly, y)

    # Generate a sequence of numbers from the minimum to the maximum of the feature
    X_seq = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)

    # Create polynomial features for the sequence
    X_seq_poly = poly.transform(X_seq)

    # Predict using the model
    y_seq_pred = model.predict(X_seq_poly)

    # Plot the data points and the fitted curve
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data points')
    plt.plot(X_seq, y_seq_pred, color='red', label='Fitted line')
    plt.xlabel(f'{feature} Ad Expenses')
    plt.ylabel('Sales')
    plt.title(f'Polynomial Regression Fit for {feature}')
    plt.legend()
    plt.show()

# Plot polynomial regression for each feature
plot_polynomial_regression('TV', degree=2)
plot_polynomial_regression('radio', degree=2)
plot_polynomial_regression('newspaper', degree=2)
```

```python
from sklearn.linear_model import Ridge
poly1 = PolynomialFeatures(degree=2)
X_poly1 = poly1.fit_transform(X)

# Split the Dataset
X_train, X_test, y_train, y_test = train_test_split(X_poly1, Y, test_size=0.2, random_state=42)

# Train the Ridge Regression Model
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

# Predict
y_pred_ridge = ridge.predict(X_test)

# Evaluate the Model
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f'Ridge Regression - Mean Squared Error: {mse_ridge}')
print(f'Ridge Regression - R^2 Score: {r2_ridge}')

# Plot Actual vs Predicted for Ridge
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_ridge, color='blue', label='Data points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Fitted line')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales (Ridge Regression)')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso

# Train the Lasso Regression Model
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

# Predict
y_pred_lasso = lasso.predict(X_test)

# Evaluate the Model
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f'Lasso Regression - Mean Squared Error: {mse_lasso}')
print(f'Lasso Regression - R^2 Score: {r2_lasso}')

# Plot Actual vs Predicted for Lasso
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_lasso, color='blue', label='Data points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Fitted line')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales (Lasso Regression)')
plt.legend()
plt.show()
```
